```python
import pandas as pd 
csv_file = 'orders.csv'
order_data = pd.read_csv(csv_file)
order_data.head()
```

```python

order_data['DeliveryTime'] = (pd.to_datetime(order_data['DeliveryDate']) - pd.to_datetime(order_data['OrderDate']))
order_data['DeliveryTime'] = order_data['DeliveryTime'].dt.days
order_data.head()
```

```python
 
csv_file = 'regions.csv'
region_data = pd.read_csv(csv_file)
region_data = region_data.drop(columns=['Area Manager','Regional Manager'])
region_data.head()
```

```python
# Convert dictionaries to DataFrames

# Merge the DataFrames on 'Territory'
merged = pd.merge(order_data, region_data, on='Territory', how='left')

merged.head()
```

```python
csv_file = 'customers.csv'
cust_data = pd.read_csv(csv_file)
cust_data['CustomerID'] = cust_data['Account Code'] 
cust_data = cust_data.drop(columns=['Key Account Manager','Account Name','Account Manager','Territory'])

cust_data.head()
```

```python
# Convert dictionaries to DataFrames

# Merge the DataFrames on 'Territory'
merged_orders = pd.merge(merged, cust_data, on='CustomerID', how='left')

merged_orders.head()
```

```python
# Save the DataFrame with more options
merged_orders.to_csv('orders_data.csv', sep=',', index=False, header=True, encoding='utf-8')
```

```python
# Convert final data to json
csv_file_path = 'orders_data.csv'
json_file_path = 'orders_data.json'

# Read the CSV data
with open(csv_file_path, mode='r') as csvfile:
    csv_reader = csv.DictReader(csvfile, delimiter=',')
    data = list(csv_reader)

# Convert to JSON
with open(json_file_path, mode='w', encoding='utf-8') as jsonfile:
    jsonfile.write(json.dumps(data))
```
